# Gaussian Mixture

```python
def optimize_gmm_n_components(X, max_components=20, criterion='bic'):
    """
    Optimize the number of components for a Gaussian Mixture Model using BIC or AIC.
    
    Args:
        X (np.ndarray): The data to fit the GMM to.
        max_components (int): The maximum number of components to try.
        criterion (str): The criterion to use for optimization ('bic' or 'aic').
    
    Returns:
        best_n_components (int): The optimal number of components.
        gmm (GaussianMixture): The GMM fitted with the optimal number of components.
    """
    bics = []
    aics = []
    n_components_range = range(1, max_components + 1)

    for n_components in n_components_range:
        gmm = GaussianMixture(n_components=n_components, random_state=42)
        gmm.fit(X)
        bics.append(gmm.bic(X))
        aics.append(gmm.aic(X))

    if criterion == 'bic':
        best_n_components = n_components_range[np.argmin(bics)]
    elif criterion == 'aic':
        best_n_components = n_components_range[np.argmin(aics)]
    else:
        raise ValueError("Criterion should be either 'bic' or 'aic'")

    best_gmm = GaussianMixture(n_components=best_n_components, random_state=42)
    best_gmm.fit(X)

    # Plot BIC/AIC scores
    plt.figure(figsize=(10, 6))
    plt.plot(n_components_range, bics, label='BIC', marker='o')
    plt.plot(n_components_range, aics, label='AIC', marker='o')
    plt.xlabel('Number of Components')
    plt.ylabel('Score')
    plt.title('BIC and AIC for GMM')
    plt.legend()
    plt.show()
```

```python
optimize_gmm_n_components(cust_prepoc)
```
